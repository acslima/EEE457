# Visualização de carregamento máximo de uma Linha de Transmissão Aérea

Emprego de fasores para visualizar os limites de carregamento

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Arc
import ipywidgets as widgets
from ipywidgets import interact

# Parameters
z = 0.1018 + 1j * 0.4894
y = 3.3753e-6
Gamma = np.sqrt(z * y)
Zc = np.sqrt(z / y)
Vn = 230e3
Pn = Vn**2 / np.real(Zc)
inom = 640  # corrente nominal do condutor

def qLT(z, y, x, Vb, Vr=1, pn=1, fp=1, i=-1):
    Gamma = np.sqrt(z * y)
    Zc = np.sqrt(z / y)
    p = Vb**2 / np.real(Zc)
    
    a = np.cosh(Gamma * x)
    b = Zc * np.sinh(Gamma * x)
    c = (1/Zc) * np.sinh(Gamma * x)
    
    vr = Vr * Vb
    sr = p / fp
    sr = sr * np.exp(1j * i * np.arccos(fp))
    ir = np.conjugate(sr) / (np.sqrt(3) * vr)
    
    vs = a * vr + b * ir
    is_ = c * vr + a * ir
    
    vsr, vsi = np.real(vs)/Vb, np.imag(vs)/Vb
    isr, isi = np.real(is_)/inom, np.imag(is_)/inom
    
    # Plot
    fig, ax = plt.subplots(figsize=(6,6))
    ax.set_aspect('equal')
    ax.set_xlim( 0.0, 1.5)
    ax.set_ylim(-1.0, 1.0)

    # Circles (arcs)
    ang = 45 # limite dos arcos em graus
    arc1 = Arc((0,0), 1.8, 1.8, theta1=-ang, theta2=ang, color='red', lw=2)
    arc2 = Arc((0,0), 2.2, 2.2, theta1=-ang, theta2=ang, color='red', lw=2)
    arc3 = Arc((0,0), 1.9, 1.9, theta1=-ang, theta2=ang, color='red', lw=2, linestyle='--')
    arc4 = Arc((0,0), 2.1, 2.1, theta1=-ang, theta2=ang, color='red', lw=2, linestyle='--')
    
    ax.add_patch(arc1)
    ax.add_patch(arc2)
    ax.add_patch(arc3)
    ax.add_patch(arc4)

    # Vectors
    ax.arrow(0, 0, vsr, vsi, head_width=0.03, head_length=0.05, fc='blue', ec='blue', lw=2)
    ax.arrow(0, 0, isr, isi, head_width=0.03, head_length=0.05, fc='green', ec='green', lw=2)

    ax.grid(True, linestyle="--", alpha=0.5)
    plt.show()

# Interactive sliders (like Mathematica Manipulate)
@interact(
    x=widgets.IntSlider(value=160, min=120, max=200, step=5, description="x"),
    vr=widgets.FloatSlider(value=1.0, min=0.9, max=1.1, step=0.05, description="Vr"),
    pn=widgets.FloatSlider(value=1.0, min=0.9, max=3.0, step=0.1, description="pn")
)
def update_plot(x, vr, pn):
    qLT(z, y, x, Vn, vr, pn)
```
