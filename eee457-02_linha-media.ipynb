Considere um circuito de 230kV com condutor ACSR (Drake), cabos para-raios de 3/8" EHS. 
Os circuitos são baseados em 230CSH1, 230CST1, 230CST3.

```python
# carregando as informações dos parâmetros unitários para cada circuito  
Vbase = 230e3
# circuito 1: 230CSH1 
r11 = 0.0820
x11 = 0.5012
b11 = 3.3180e-6
zc1 = np.sqrt( x11/ b11 )
Pn1 = Vbase**2 * np.sqrt(b11/x11)

# circuito 2: 230CST1
r12 = 0.0815
x12 = 0.4402
b12 = 3.7504e-6
Pn2 = Vbase**2 * np.sqrt(b12/x12)

# circuito 3: 230CST3
r13 = 0.0818
x13 = 0.4812
b13 = 3.43760e-6
Pn3 = Vbase**2 * np.sqrt(b13/x13)
```
